1.  **`text_to_speak`**: We write the message we want the computer to say here.
2.  **`gTTS()`**: This takes our text and creates the actual speech sound.
3.  **`tts.save()`**: This saves the generated speech as an audio file (an MP3).
4.  **`Audio()`**: This plays the saved audio file right here in Colab so you can hear it.

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text_to_speak = "Hello! I am your own voice assistant"
tts = gTTS(text=text_to_speak, lang='en')
tts.save("voice_assistant.mp3")
Audio("voice_assistant.mp3", autoplay=True)
```

```python
import google.generativeai as genai

# Configure your API key
# Replace 'YOUR_API_KEY' with your actual API key
genai.configure(api_key='')

# Initialize the Generative Model (using a free model like 'gemini-pro')
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
prompt = "What capital of Malaysia"
response = model.generate_content(prompt)
print(response.text)
```

```python
user_input = input("Enter your prompt for the AI ")
llm_prompt = f"Give a humorous and extremely brief answer (maximum of 2-3 sentences) to the following: {user_input}"
response = model.generate_content(llm_prompt)
print(response.text)
```

```python
text_to_speak_from_llm = response.text
tts_llm = gTTS(text=text_to_speak_from_llm, lang='en')
tts_llm.save("llm_response.mp3")
Audio("llm_response.mp3", autoplay=True)
```
